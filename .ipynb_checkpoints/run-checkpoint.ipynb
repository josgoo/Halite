```python
def onLine(x1,y1,x2,y2, tx,ty):
    m = (y2 - y1) / (x2 - x1) if x2 != x1 else None
    b = y1 - x1 * m
    return (m * tx + b == ty) if x2 != x1 else x1 == tx
    
```

```python
onLine(0,0,2,6,4,8)
```
